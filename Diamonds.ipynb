```python
# this line is a comment line # sign turns lines into comments

# import libraries, after improting you can use pd for pandas or np for numpy
import pandas as pd
import numpy as np
import matplotlib.pyplot as pl
```

```python
d = pd.read_csv('Pandas Cookbook/data/diamonds.csv')
d['vol'] = pd.Series(d['x'] * d['y'] * d['z'], index=d.index)
d = d.drop(['x','y','z'], axis=1)
d.corr()
```

```python
d.head()
```

```python
clarity = {'clarity' : {'IF' : 9, 'VVS1' : 8, 'VVS2' : 7, 'VS1' : 6, 'VS2' : 5, 'SI1' : 4, 'SI2' : 3, 'I1' : 2, 'I2' : 1}}
cut = {'cut' : {'Ideal' : 5, 'Premium' : 4, 'Very Good' : 3, 'Good' : 2, 'Fair' : 1}}
color = {'color' : {'D' : 7, 'E' : 6, 'F' : 5, 'G' : 4, 'H' : 3, 'I' : 2, 'J' : 1}}
d.replace(clarity, inplace=True)
d.replace(cut, inplace=True)
d.replace(color, inplace=True)
```

```python
d.head()
```

```python
d.corr()
```

```python
d.groupby('color')['price'].mean() / d.groupby('color')['cut'].mean() / d.groupby('color')['carat'].mean()
```

```python
d['clarity'] = d['clarity']/9
pl.figure()
d.plot.scatter(x='carat',y='price',c='clarity',s=d['cut']*5,loglog=True, legend=True,figsize=(11,8))
pl.show()
```

```python
d.groupby('cut')['price'].median()
```

```python
d.groupby('clarity')['price'].median()
```
